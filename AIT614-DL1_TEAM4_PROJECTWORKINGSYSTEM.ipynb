```python
data_clean["Water_Quality"] = np.where(data_clean['pH (standard units)'] > 7, 1, 0)
data_clean
#data_clean.dtypes
```

```python
data_clean = data_clean.drop(['pH (standard units)'], axis=1)
data_clean = data_clean.drop(['Site_Id'], axis=1)
```

```python
data_clean.count()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from scipy.stats import randint
from sklearn.metrics import f1_score
```

```python

# Split the data into features (X) and target (y)
X = data_clean.drop('Water_Quality', axis=1)
y = data_clean['Water_Quality']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
X.count()
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

cm = confusion_matrix(y_test, y_pred)

ConfusionMatrixDisplay(confusion_matrix=cm).plot();

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
#f1_score = f1_score(y_test, y_pred)
f1_score = 2*((precision * recall)/(precision + recall))

print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1_score)
```

```python

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Create the confusion matrix
cm = confusion_matrix(y_test, y_pred)

ConfusionMatrixDisplay(confusion_matrix=cm).plot();

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
#f1_score = f1_score(y_test, y_pred)
f1_score = 2*((precision * recall)/(precision + recall))

print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1_score)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

y_pred = dt.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

cm = confusion_matrix(y_test, y_pred)

ConfusionMatrixDisplay(confusion_matrix=cm).plot();

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
#f1_score = f1_score(y_test, y_pred)
f1_score = 2*((precision * recall)/(precision + recall))

print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1_score)
```
